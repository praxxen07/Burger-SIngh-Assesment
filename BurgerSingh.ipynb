```python
start_date='2023-09-01'
end_date='2023-10-31'

df1= df1[(df1['Business_Date']>=start_date) & (df1['Business_Date']<=end_date)]
df2= df2[(df2['Business_Date']>=start_date) & (df2['Business_Date']<=end_date)]

print(df1.shape)
print(df2.shape)
```

```python
## Calculate sales by channel
sales_by_channel = df1.groupby("Order_Source")["Gross_Amount"].sum()

swiggy_sales = sales_by_channel.get("swiggy", 0)
zomato_sales = sales_by_channel.get("zomato", 0)

# Apply commissions
swiggy_commission = swiggy_sales * 0.223
zomato_commission = zomato_sales * 0.237
total_commission = swiggy_commission + zomato_commission

## Calculate totals
total_revenue = df1["Gross_Amount"].sum()
total_discounts = df1["Discounts"].sum()
total_net_sales = df1["Net_Amount"].sum()  # after discounts but before tax
total_taxes = df1["Taxes"].sum()
total_cogs = df1["Materials_Cost"].sum() + df1["Supplies_Cost"].sum()     #  COGS (Cost of Goods Sold)

## Calculate profits
gross_profit = total_revenue - total_cogs - total_commission
net_profit = gross_profit - total_taxes

## Create P&L DataFrame
pnl_data = {
    "Total Revenue (Gross Sales)": [total_revenue],
    "Discounts": [total_discounts],
    "Net Sales (before tax)": [total_net_sales],
    "Swiggy Commission": [swiggy_commission],
    "Zomato Commission": [zomato_commission],
    "Total Commission": [total_commission],
    "COGS (Materials + Supplies)": [total_cogs],
    "Gross Profit": [gross_profit],
    "Taxes": [total_taxes],
    "Net Profit": [net_profit]
}

pnl_df = pd.DataFrame(pnl_data)
print("\n Profit & Loss Statement (Sept & Oct 2023) :")
pnl_df
```

### QUESTION 2:

```python
pnl_df["Gross Margin %"] = (pnl_df["Gross Profit"] / pnl_df["Total Revenue (Gross Sales)"]) * 100
pnl_df["Net Margin %"] = (pnl_df["Net Profit"] / pnl_df["Total Revenue (Gross Sales)"]) * 100
pnl_df["COGS %"] = (pnl_df["COGS (Materials + Supplies)"] / pnl_df["Total Revenue (Gross Sales)"]) * 100
pnl_df["Commission %"] = (pnl_df["Total Commission"] / pnl_df["Total Revenue (Gross Sales)"]) * 100

print("\nPerformance Metrics:")
Performance_df= pd.DataFrame(pnl_df[["Gross Margin %", "Net Margin %", "COGS %", "Commission %"]])
Performance_df
```

```python
total_orders = df2["Statement_Number"].nunique()
aov = total_revenue / total_orders  # Average Order Value

# Orders by channel
orders_by_channel = df2["Channel"].value_counts().reset_index()
orders_by_channel.columns = ["Channel", "Number of Orders"]

# 3. Revenue by category
category_sales = (
    df2.groupby("Category")["Gross_Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Gross_Amount", ascending=False)
)

# Top 5 best-selling items
top_items = (
    df2.groupby("Item_Name")["Gross_Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Gross_Amount", ascending=False)
    .head(5)
)

# Bottom 5 least-selling items
bottom_items = (
    df2.groupby("Item_Name")["Gross_Amount"]
    .sum()
    .reset_index()
    .sort_values(by="Gross_Amount", ascending=True)
    .head(5)
)

print("\n--- Additional Operational Metrics (Q2 Enrichment) ---")
print(f"Total Orders: {total_orders}")
print(f"Average Order Value (AOV): ₹{aov:,.2f}\n")

print("Orders by Channel:")
print(orders_by_channel.to_string(index=False), "\n")

print("Revenue by Category:")
print(category_sales.to_string(index=False), "\n")

print("Top 5 Best-Selling Items:")
print(top_items.to_string(index=False), "\n")

print("Bottom 5 Least-Selling Items:")
print(bottom_items.to_string(index=False))
```

```python
df1['Month'] = df1['Business_Date'].dt.strftime('%B')

# Separate Swiggy and Zomato for commissions
swiggy_df = df1[df1['Order_Source'].str.lower() == 'swiggy']
zomato_df = df1[df1['Order_Source'].str.lower() == 'zomato']

# Group by month and calculate P&L components
pl_monthly = df1.groupby('Month').apply(lambda x: pd.Series({
    'Total Revenue (Gross Sales)': x['Gross_Amount'].sum(),
    'Discounts': x['Discounts'].sum(),
    'Net Sales (before tax)': x['Net_Amount'].sum(),
    'Swiggy Commission': x.loc[x['Order_Source'].str.lower() == 'swiggy', 'Gross_Amount'].sum() * 0.223,
    'Zomato Commission': x.loc[x['Order_Source'].str.lower() == 'zomato', 'Gross_Amount'].sum() * 0.237,
    'COGS': x['Materials_Cost'].sum() + x['Supplies_Cost'].sum(),
    'Taxes': x['Taxes'].sum()
})).reset_index()

pl_monthly['Total Commission'] = pl_monthly['Swiggy Commission'] + pl_monthly['Zomato Commission']
pl_monthly['Gross Profit'] = pl_monthly['Net Sales (before tax)'] - pl_monthly['Total Commission'] - pl_monthly['COGS']
pl_monthly['Net Profit'] = pl_monthly['Gross Profit'] - pl_monthly['Taxes']
pl_monthly= pd.DataFrame(pl_monthly)
pl_monthly
```
